Creating csv file with data found here: https://cvml.ist.ac.at/AwA2/

```python
import pandas as pd
import numpy as np
```

```python
#Loading files
animals = pd.read_csv('classes.txt', sep="\t", header=None)
attributes = pd.read_csv('predicate-matrix-binary.txt', sep=" ", header=None)
attribute_names = pd.read_csv('predicates.txt', sep="\t", header=None)
```

```python
animals = animals.drop(columns = [0])
attribute_names = attribute_names.drop(columns= [0])
```

```python
animals = np.asarray(animals[1])
animals = {number: value for number, value in enumerate(animals)}
```

```python
attribute_names = np.asarray(attribute_names[1])
attribute_names = {number: value for number, value in enumerate(attribute_names)}
```

```python
attribute_names
```

```python
final = attributes.rename(columns=attribute_names, index=animals)
```

```python
final.to_csv('AwA.csv')
```
